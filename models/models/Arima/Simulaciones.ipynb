## CoÃÅdigo para simula un ruido blanco

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima_process import ArmaProcess
```

```python
# Generar 100 n√∫meros aleatorios con media 0 y varianza 1
g = np.random.normal(0, 1, 100)

# Configurar el tama√±o de la figura
plt.figure(figsize=(12, 5))

# Graficar la serie temporal
plt.subplot(1, 2, 1)
plt.plot(g)
plt.title('Nii(0,1)')
plt.xlabel('Tiempo')
plt.ylabel('Valor')

# Graficar la funci√≥n de autocorrelaci√≥n
plt.subplot(1, 2, 2)
plot_acf(g, lags=20, ax=plt.gca(), alpha=0.05)
plt.ylim(-1, 1)
plt.title('Autocorrelaci√≥n Simple')

# Mostrar los gr√°ficos
plt.tight_layout()
plt.show()
```

## CoÃÅdigo para simular un camino aleatorio

```python
# Generar 100 n√∫meros aleatorios con media 0 y varianza 1
w = np.random.normal(0, 1, 100)

# Inicializar el camino aleatorio
x = np.zeros(100)
x[0] = w[0]

# Simular el camino aleatorio
for t in range(1, 100):
    x[t] = x[t - 1] + w[t]

# Calcular la primera diferencia
d = np.diff(x)

# Configurar el tama√±o de la figura
plt.figure(figsize=(12, 10))

# Graficar el camino aleatorio
plt.subplot(2, 2, 1)
plt.plot(x)
plt.title('Camino aleatorio Xt')
plt.xlabel('Tiempo')
plt.ylabel('Valor')

# Graficar la funci√≥n de autocorrelaci√≥n del camino aleatorio
plt.subplot(2, 2, 2)
plot_acf(x, lags=20, ax=plt.gca(), alpha=0.05)
plt.ylim(-0.5, 0.5)
plt.title('Autocorrelaci√≥n Simple de Xt')

# Graficar la primera diferencia del camino aleatorio
plt.subplot(2, 2, 3)
plt.plot(d)
plt.title('Primera diferencia de Xt')
plt.xlabel('Tiempo')
plt.ylabel('Valor')

# Graficar la funci√≥n de autocorrelaci√≥n de la primera diferencia
plt.subplot(2, 2, 4)
plot_acf(d, lags=20, ax=plt.gca(), alpha=0.05)
plt.ylim(-0.5, 0.5)
plt.title('Autocorrelaci√≥n Simple de d')

# Ajustar el dise√±o de los gr√°ficos y mostrar
plt.tight_layout()
plt.show()
```

## CoÃÅdigo para simular un proceso auroregresivo de orden 1 AR(1)

```python
# Definir los par√°metros del proceso AR(1)
ar_params = np.array([1, -0.4])  # Note que AR params deben ser negados
ma_params = np.array([1])  # No hay componente MA

# Crear el proceso AR(1)
ar_process = ArmaProcess(ar_params, ma_params)

# Generar 100 observaciones del proceso AR(1)
AR = ar_process.generate_sample(nsample=100)

# Configurar el tama√±o de la figura y el layout
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(AR)
plt.title(r'AR(1) $\phi$=0.4')
plt.xlabel('Tiempo')
plt.ylabel('Valor')

# Graficar la funci√≥n de autocorrelaci√≥n simple
plt.subplot(2, 2, 2)
plot_acf(AR, lags=20, ax=plt.gca(), alpha=0.05)
plt.ylim(-0.5, 0.5)
plt.title('Autocorrelaci√≥n Simple')

# Graficar la funci√≥n de autocorrelaci√≥n parcial
plt.subplot(2, 2, 3)
plot_pacf(AR, lags=20, ax=plt.gca(), alpha=0.05)
plt.ylim(-0.5, 0.5)
plt.title('Autocorrelaci√≥n Parcial')

# Ajustar el dise√±o de los gr√°ficos y mostrar
plt.tight_layout()
plt.show()
```

# CoÃÅdigo para simular un proceso de media moÃÅvil de orden 1 MA(1)

```python
# Definir los par√°metros del proceso MA(1)
ar_params = np.array([1])  # No hay componente AR
ma_params = np.array([1, 0.5])

# Crear el proceso MA(1)
ma_process = ArmaProcess(ar_params, ma_params)

# Generar 100 observaciones del proceso MA(1)
MA = ma_process.generate_sample(nsample=100)

# Configurar el tama√±o de la figura y el layout
plt.figure(figsize=(12, 8))

# Graficar la serie temporal del proceso MA(1)
plt.subplot(2, 2, 1)
plt.plot(MA)
plt.title(r'MA(1) $\theta=0.5$')
plt.xlabel('Tiempo')
plt.ylabel('Valor')

# Graficar la funci√≥n de autocorrelaci√≥n simple
plt.subplot(2, 2, 2)
plot_acf(MA, lags=20, ax=plt.gca(), alpha=0.05)
plt.ylim(-0.5, 0.5)
plt.title('Autocorrelaci√≥n Simple')

# Graficar la funci√≥n de autocorrelaci√≥n parcial
plt.subplot(2, 2, 3)
plot_pacf(MA, lags=20, ax=plt.gca(), alpha=0.05)
plt.ylim(-0.5, 0.5)
plt.title('Autocorrelaci√≥n Parcial')

# Ajustar el dise√±o de los gr√°ficos y mostrar
plt.tight_layout()
plt.show()
```
